```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('./merged-data.csv')
values_to_replace = [9999.99, 99999.9, 999.999, 9999999.0]
df.replace(values_to_replace, np.nan, inplace=True)
```

```python
plt.scatter(df['year'],df['intensity'])
plt.xticks(range(int(df['year'].min()), int(df['year'].max()) + 1, 1))
plt.xlabel('year')
plt.ylabel('intensity')
```

```python
plt.scatter(df['month'],df['intensity'])
plt.xlabel('month')
plt.ylabel('intensity')
```

```python
plt.scatter(df['month'],df['bt'])
plt.xlabel('month')
plt.ylabel('bt')
```
